```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
def preprocess(img, imgsize):
    # there are damaged files in iam dataset - just use black image instead
    if img is None:
        img = np.zeros([1, 1])

    # create target image and copy sample image into it
    (wt, ht) = imgsize
    (h, w) = img.shape
    fx = w / wt
    fy = h / ht
    f = max(fx, fy)
    newsize = (min(wt, int(w / f)), min(ht, int(h / f)))
    img = cv2.resize(img, newsize)
    target = np.ones([ht, wt]) * 255
    target[0:newsize[1], 0:newsize[0]] = img

    # transpose for tf
    img = cv2.transpose(target)
    
    # normalize
    (m, s) = cv2.meanStdDev(img)
    m=m[0][0]
    s=s[0][0]
    img = img - m
    img = img / s if s>0 else img
    return img
```

### Original Image

```python
imgFile = "../data/test.png"
img = cv2.imread(imgFile, cv2.IMREAD_GRAYSCALE)

print(img.shape)
plt.imshow(img)
```

### Processed Image

```python
imgsize = (128, 32)
img = preprocess(img, imgsize)
print(img.shape)
plt.imshow(img)
```

## How it works?

```python
imgFile = "../data/test.png"
img = cv2.imread(imgFile, cv2.IMREAD_GRAYSCALE)
```

```python
# initial constants
imgsize = (128, 32) # wanted size
wt, ht = imgsize
h, w = img.shape
```

```python
# newsize: scaling original img into wanted size
fx, fy = w/wt, h/ht
f = max(fx, fy)
newsize = (min(wt, int(w/f)), min(ht, int(h/f)))
print('newsize: ', newsize)

img = cv2.resize(img, newsize)
print('img.shape: ', img.shape)
plt.imshow(img)
```

```python
# target: making white background sized (32, 128)
target = np.ones([ht, wt]) * 255
print(target.shape)
plt.imshow(target)
```

```python
# put processed img into background target
target[0:newsize[1], 0:newsize[0]] = img
plt.imshow(target)
```

```python
# transpose it for CNN
img = cv2.transpose(target)
print(img.shape)
plt.imshow(img)
```
